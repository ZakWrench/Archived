```python
#Selection by label
print(df_sub.loc[:,'Name'])
print(df_sub.loc[:2,['Name','Physics score']])
```

```python
#Setting and filtering values
#Change Jenny's name to Fatihi

df_sub.loc[df_sub['Name'] == 'Jenny', 'Name'] = 'Fatihi'
print(df_sub)

df_sub.iloc[0,0] = 'Fatihi2'
print(df_sub)

#Physics score more than equal to 90
print(df_sub.loc[df_sub['Physics score'] >= 90,['Name','Physics score']])
```

```python
#Missing value treatment
df_sub.iloc[0,2] = np.nan
print(df_sub)

df_sub.dropna(how='any') #drop any rows that have missing data.
```

```python
#Substitute missing value with mean
mean_score = np.mean(df_sub['Physics score'])
df_sub.fillna(value=mean_score)
```

```python
#Merging dataframe
df_concat = pd.concat([df_sub, new_df])
df_concat.reset_index(inplace = True)
df_concat.drop(columns = ['index'],inplace = True)
df_concat
```

```python
#Join
s1 = ['Jerry', 'Ravi', 'John', 'Harry', 'Rachel']
s2 = ['A', 'B', 'C', 'B', 'A']
df_last_name = pd.DataFrame({'Name':s1, 'Section':s2})
df_last_name

#pd.merge(left, right, on = 'key', how = 'inner')
df_merge = pd.merge(df_concat, df_last_name, on = 'Name', how = 'inner')
df_merge
```

```python
#Grouping data
#average score in math per section

print(df_merge.groupby(['Section'])['Math score'].mean())
print(df_merge.groupby(['Section'])['Name'].count())
```

```python
#Loops in pandas
for index, row in df_merge.iterrows():
    print(index) #index of df
    print(row['Name'])#print row of df
    print(df_merge.loc[index,'Math score'])
```

```python
#functions in pandas
def rating_function(x):
    if x >= 80:
        return "good"
    elif (x < 80) & (x > 60):
        return "descent"
    else:
        return "bad"
```

```python
#rating_function(df_merge['Math score'])
for index, row in df_merge.iterrows():
    df_merge.loc[index,'Math grade'] = rating_function(row['Math score'])

df_merge
```

```python
#apply fn
df_merge['Math grade'] = df_merge['Math score'].apply(rating_function)
df_merge
```
